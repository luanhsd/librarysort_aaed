An√°lise de Desempenho do Algoritmo Library Sort

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result/librarysort.csv')
df.rename(columns={'compare': 'compares'}, inplace=True)
df = df.sort_values(by=['quantity'])
print(df)
```

Resultados obtidos na execucao do algoritmo Library Sort com os quatro tipos de conjuntos definidos no cenario.

```python
plt.plot(df[df.order == 'ORDERED'].quantity,
         df[df.order == 'ORDERED'].processing_time, label = "Conjunto Ordenado")
plt.plot(df[df.order == 'INVERSE'].quantity,
         df[df.order == 'INVERSE'].processing_time, label="Conjunto Ordenado Inversamente")
plt.plot(df[df.order == 'ALMOST'].quantity,
         df[df.order == 'ALMOST'].processing_time, label="Conjunto Quase Ordenado")
plt.plot(df[df.order == 'RANDOM'].quantity,
         df[df.order == 'RANDOM'].processing_time, label="Conjunto Desordenado")
plt.title("Library Sort - Tempo de Processamento")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Tempo de Processamento")
plt.legend()
plt.savefig("processing_time_library.png")
```

Podemos inferir que o pior caso do algoritmo e quando o conjunto esta inversamente ordenados, tendo entao uma complexidade O(N¬≤).

```python
plt.plot(df[df.order == 'ORDERED'].quantity,
         df[df.order == 'ORDERED'].moves, label="Conjunto Ordenado")
plt.plot(df[df.order == 'INVERSE'].quantity,
         df[df.order == 'INVERSE'].moves, label="Conjunto Ordenado Inversamente")
plt.plot(df[df.order == 'ALMOST'].quantity,
         df[df.order == 'ALMOST'].moves, label="Conjunto Quase Ordenado")
plt.plot(df[df.order == 'RANDOM'].quantity,
         df[df.order == 'RANDOM'].moves, label="Conjunto Desordenado")
plt.title("Library Sort - Quantidade de Movimentacoes")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Quantidade de Movimentacoes")
plt.legend()
plt.savefig("moves_library.png")
```

Quando o conjunto esta inversamente ordenado o algoritmo realiza uma quandidade enorme de movimentacoes passando por todo o conjunto e depois removendo os espacos em branco.

```python
plt.plot(df[df.order == 'ORDERED'].quantity,
         df[df.order == 'ORDERED'].compares, label="Conjunto Ordenado")
plt.plot(df[df.order == 'INVERSE'].quantity,
         df[df.order == 'INVERSE'].compares, label="Conjunto Ordenado Inversamente")
plt.plot(df[df.order == 'ALMOST'].quantity,
         df[df.order == 'ALMOST'].compares, label="Conjunto Quase Ordenado")
plt.plot(df[df.order == 'RANDOM'].quantity,
         df[df.order == 'RANDOM'].compares, label="Conjunto Desordenado")
plt.title("Library Sort - Quantidade de Comparacoes")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Quantidade de Comparacoes")
plt.legend()
plt.savefig("compares_library.png")
```

```python
df = pd.read_csv('result/result.csv')
df.rename(columns={'compare': 'compares'}, inplace=True)
df = df.sort_values(by=['quantity'])
print(df)
```

```python
ordered = df[df.order == 'ORDERED']
plt.plot(ordered[ordered.algorithm == 'bubblesort'].quantity,
        ordered[ordered.algorithm == 'bubblesort'].processing_time, label="Bubble Sort")
plt.plot(ordered[ordered.algorithm == 'insertionsort'].quantity,
        ordered[ordered.algorithm == 'insertionsort'].processing_time, label="Insertion Sort")
plt.plot(ordered[ordered.algorithm == 'selectionsort'].quantity,
        ordered[ordered.algorithm == 'selectionsort'].processing_time, label="Selection Sort")
plt.plot(ordered[ordered.algorithm == 'heapsort'].quantity,
        ordered[ordered.algorithm == 'heapsort'].processing_time, label="Heap Sort")
plt.plot(ordered[ordered.algorithm == 'mergesort'].quantity,
        ordered[ordered.algorithm == 'mergesort'].processing_time, label="Merge Sort")
plt.plot(ordered[ordered.algorithm == 'quicksort'].quantity,
        ordered[ordered.algorithm == 'quicksort'].processing_time, label="Quick Sort")
plt.plot(ordered[ordered.algorithm == 'librarysort'].quantity,
        ordered[ordered.algorithm == 'librarysort'].processing_time, label="Library Sort")

plt.title("Algoritmos de Ordenacao - Tempo de Processamento (Arranjos Ordenados)")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Tempo de Processamento")
plt.legend()
plt.savefig("processing_time_ordered.png")
```

Comparando com todos os algoritmos de ordenacao escolhidos para a analise podemos observer que, quando o conjunto esta ordenado, o Library Sort consegue ter um bom caso, tendo entao uma complexidade O(n log n). 

```python
inverse = df[df.order == 'INVERSE']
plt.plot(inverse[inverse.algorithm == 'bubblesort'].quantity,
         inverse[inverse.algorithm == 'bubblesort'].processing_time, label="Bubble Sort")
plt.plot(inverse[inverse.algorithm == 'insertionsort'].quantity,
         inverse[inverse.algorithm == 'insertionsort'].processing_time, label="Insertion Sort")
plt.plot(inverse[inverse.algorithm == 'selectionsort'].quantity,
         inverse[inverse.algorithm == 'selectionsort'].processing_time, label="Selection Sort")
plt.plot(inverse[inverse.algorithm == 'heapsort'].quantity,
         inverse[inverse.algorithm == 'heapsort'].processing_time, label="Heap Sort")
plt.plot(inverse[inverse.algorithm == 'mergesort'].quantity,
         inverse[inverse.algorithm == 'mergesort'].processing_time, label="Merge Sort")
plt.plot(inverse[inverse.algorithm == 'quicksort'].quantity,
         inverse[inverse.algorithm == 'quicksort'].processing_time, label="Quick Sort")
plt.plot(inverse[inverse.algorithm == 'librarysort'].quantity,
         inverse[inverse.algorithm == 'librarysort'].processing_time, label="Library Sort")

plt.title("Algoritmos de Ordenacao - Tempo de Processamento (Arranjos Inversamente Ordenados)")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Tempo de Processamento")
plt.legend()
plt.savefig("processing_time_inverse.png")
```

Quando executado os algoritmos com arranjos inversamente Ordenados o Library Sort tem seu pior caso mas se torna ainda mais eficiente que o Insertion Sort, o algoritmo a qual foi baseado. Isso tende a reducao na movimentacao, uma vez que o Library Sort ja possui espacos reservados para as mesmas.

```python
almost = df[df.order == 'ALMOST']
plt.plot(almost[almost.algorithm == 'bubblesort'].quantity,
         almost[almost.algorithm == 'bubblesort'].processing_time, label="Bubble Sort")
plt.plot(almost[almost.algorithm == 'insertionsort'].quantity,
         almost[almost.algorithm == 'insertionsort'].processing_time, label="Insertion Sort")
plt.plot(almost[almost.algorithm == 'selectionsort'].quantity,
         almost[almost.algorithm == 'selectionsort'].processing_time, label="Selection Sort")
plt.plot(almost[almost.algorithm == 'heapsort'].quantity,
         almost[almost.algorithm == 'heapsort'].processing_time, label="Heap Sort")
plt.plot(almost[almost.algorithm == 'mergesort'].quantity,
         almost[almost.algorithm == 'mergesort'].processing_time, label="Merge Sort")
plt.plot(almost[almost.algorithm == 'quicksort'].quantity,
         almost[almost.algorithm == 'quicksort'].processing_time, label="Quick Sort")
plt.plot(almost[almost.algorithm == 'librarysort'].quantity,
         almost[almost.algorithm == 'librarysort'].processing_time, label="Library Sort")

plt.title(
    "Algoritmos de Ordenacao - Tempo de Processamento (Arranjos Quase Ordenados)")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Tempo de Processamento")
plt.legend()
plt.savefig("processing_time_almost.png")
```

```python
almost = df[df.order == 'ALMOST']
plt.plot(almost[almost.algorithm == 'insertionsort'].quantity,
         almost[almost.algorithm == 'insertionsort'].processing_time, label="Insertion Sort")
plt.plot(almost[almost.algorithm == 'librarysort'].quantity,
         almost[almost.algorithm == 'librarysort'].processing_time, label="Library Sort")

plt.title(
    "Algoritmos de Ordenacao - Tempo de Processamento (Arranjos Quase Ordenados)")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Tempo de Processamento")
plt.legend()
plt.savefig("processing_time_almost_with_library_insertion.png")
```

O tempo de processamento do Library Sort com arranjo quase ordenado se mostrou menos eficiente apenas em comparacao com o Insertion Sort. Comparado com os outros algoritmos de ordenacao ele mostrou uma melhor performance em tempo de processamento.

```python
random = df[df.order == 'RANDOM']
plt.plot(random[random.algorithm == 'bubblesort'].quantity,
         random[random.algorithm == 'bubblesort'].processing_time, label="Bubble Sort")
plt.plot(random[random.algorithm == 'insertionsort'].quantity,
         random[random.algorithm == 'insertionsort'].processing_time, label="Insertion Sort")
plt.plot(random[random.algorithm == 'selectionsort'].quantity,
         random[random.algorithm == 'selectionsort'].processing_time, label="Selection Sort")
plt.plot(random[random.algorithm == 'heapsort'].quantity,
         random[random.algorithm == 'heapsort'].processing_time, label="Heap Sort")
plt.plot(random[random.algorithm == 'mergesort'].quantity,
         random[random.algorithm == 'mergesort'].processing_time, label="Merge Sort")
plt.plot(random[random.algorithm == 'quicksort'].quantity,
         random[random.algorithm == 'quicksort'].processing_time, label="Quick Sort")
plt.plot(random[random.algorithm == 'librarysort'].quantity,
         random[random.algorithm == 'librarysort'].processing_time, label="Library Sort")

plt.title(
    "Algoritmos de Ordenacao - Tempo de Processamento (Arranjos Nao Ordenados)")
plt.xlabel("Quantidade de Elementos")
plt.ylabel("Tempo de Processamento")
plt.legend()
plt.savefig("processing_time_random.png")
```

Com arranjos nao ordenados o Library tambem se mostrou melhor comparado com o Insertion Sort. Teve pior desempenho comparado com o Quick Sort, Merge Sort e Heap Sort.
